```python
import pandas as pd
import zipfile
import os

# Unzip the archive
zip_file_name = 'archive (6).zip'
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall('.')

# Assuming the CSVs are in a subfolder after unzipping, e.g., 'banking_dataset_kaggle/data/csv/'
# You might need to adjust this path based on the actual structure inside the zip file.
# Let's check the common path based on the previous context provided in kernel state.
# From the 'uploaded' variable: 'banking_dataset_kaggle/data/csv/accounts.csv'

data_path = 'banking_dataset_kaggle/data/csv/'

customers = pd.read_csv(os.path.join(data_path, "customers.csv"))
accounts = pd.read_csv(os.path.join(data_path, "accounts.csv"))
loans = pd.read_csv(os.path.join(data_path, "loans.csv"))
cards = pd.read_csv(os.path.join(data_path, "cards.csv"))
branches = pd.read_csv(os.path.join(data_path, "branches.csv"))
merchants = pd.read_csv(os.path.join(data_path, "merchants.csv"))
```

```python
con = duckdb.connect()

con.register("customers", customers)
con.register("accounts", accounts)
con.register("loans", loans)
con.register("cards", cards)
con.register("branches", branches)
con.register("merchants", merchants)
```
